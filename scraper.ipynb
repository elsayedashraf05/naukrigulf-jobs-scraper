# 📌 Code Explanation – Naukrigulf Job Scraper

This code uses **Python, Selenium, and Pandas** to scrape **Data Engineer job listings** from the **Naukrigulf** website and collect job details from the first **3 pages** of search results.

---

## 🔹 Libraries Used
- **selenium**: To automate browser actions and interact with web elements.
- **pandas**: To organize the extracted data into a structured format.
- **time**: Used to add delays to ensure pages load properly.

---

## 🔹 Selenium WebDriver Setup
The Chrome browser is configured with options to reduce automation detection by disabling Selenium flags and extensions.  
The browser window is maximized for better element visibility.

---

## 🔹 Opening the Target Website
The script opens the Naukrigulf website filtered for **Data Engineer jobs**.

---

## 🔹 Pagination Handling
The script loops through the **first 3 pages** of job results.  
On each page, it waits for the page to load before scraping the job listings.

---

## 🔹 Job Data Extraction
For each job card:
- **Job Title** is extracted from the job title element.
- **Company Name** is extracted from the company name section.
- **Experience Required** is extracted from the experience field.
- **Job Location** is extracted from the location field.
- **Job Description** is collected from the short description shown on the listing page.

Each job’s data is stored as a dictionary and appended to a list.

---

## 🔹 Error Handling
A `try-except` block is used to:
- Skip jobs with missing data

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd

list_jobs = []

options = Options()
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option("useAutomationExtension", False)

driver = webdriver.Chrome(options=options)
driver.maximize_window()
driver.get("https://www.naukrigulf.com/data-engineer-jobs")

for page in range(3):
    time.sleep(3)

    jobs = driver.find_elements(By.CLASS_NAME, "srp-tuple")

    for job in jobs:
        try:
            title = job.find_element(By.CLASS_NAME, "designation-title").text
            company = job.find_element(By.CLASS_NAME, "info-org").text
            experience = job.find_element(
                By.CSS_SELECTOR,
                "li.info-exp span:nth-child(2)"
            ).text
            location = job.find_element(
                By.CSS_SELECTOR,
                "li.info-loc span:nth-child(2)"
            ).text
            description = job.find_element(By.CLASS_NAME, "description").text

            list_jobs.append({
                "title": title,
                "company": company,
                "experience": experience,
                "location": location,
                "description": description
            })

        except Exception as e:
            print("Skipped one job:", e)

    # go to next page (only if not last page)
    if page < 2:
        try:
            next_btn = driver.find_element(By.XPATH, "//span[@class='ico bwd']")
            driver.execute_script("arguments[0].click();", next_btn)
        except:
            print("Next page button not found")
            break

driver.quit()
```

```python
df = pd.DataFrame(list_jobs)
df
```

```python
df.to_csv("Jobs.csv", index=False)
```
